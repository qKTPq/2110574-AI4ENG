# Task
Analyze the "winequality_red.csv" dataset by performing a linear regression analysis to predict the 'quality' variable. The analysis should include checking the linearity assumption, building and evaluating a linear regression model using a train-test split (test_size=0.25, random_state=101), reporting the model equation, checking for out-of-range predictions, and improving the model through post-processing, followed by a re-evaluation of the model's performance.

## Load the data

### Subtask:
Load the `winequality_red.csv` file into a pandas DataFrame.


**Reasoning**:
Import pandas and load the csv file into a dataframe, then display the head and info of the dataframe to verify the loading process.



```python
import pandas as pd

df = pd.read_csv('winequality_red.csv')

display(df.head())
display(df.info())
```

**Reasoning**:
The data is not loaded correctly as it's in a single column. The separator needs to be specified as ';'. Reload the data with the correct separator and display the head and info again.



```python
df = pd.read_csv('winequality_red.csv', sep=';')

display(df.head())
display(df.info())
```

## Check linearity assumption

### Subtask:
Calculate the correlation between each feature and the target variable (`quality`).


**Reasoning**:
Calculate the correlation matrix and extract the correlations with the 'quality' column.



```python
correlation_matrix = df.corr()
quality_correlation = correlation_matrix['quality'].sort_values(ascending=False)
display(quality_correlation)
```

## Prepare data for modeling

### Subtask:
Split the data into features (X) and target (y), and then split these into training and testing sets using `train_test_split` with `test_size=0.25` and `random_state=101`.


**Reasoning**:
Split the data into features (X) and target (y), and then split these into training and testing sets using train_test_split with test_size=0.25 and random_state=101.



```python
from sklearn.model_selection import train_test_split

X = df.drop('quality', axis=1)
y = df['quality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

## Create and train the model

### Subtask:
Build a linear regression model using the selected features and train it on the training data.


**Reasoning**:
Build and train a linear regression model using the training data.



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

## Report the equation

### Subtask:
Display the coefficients and intercept of the trained linear model.


**Reasoning**:
Access and display the coefficients and intercept of the trained linear regression model.



```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

## Predict on the test set

### Subtask:
Make predictions on the test set using the trained model.


**Reasoning**:
Use the trained linear regression model to make predictions on the test feature set.



```python
predictions = model.predict(X_test)
display(predictions)
```

## Evaluate the model

### Subtask:
Calculate and report the RMSE and R-squared score on the test set.


**Reasoning**:
Calculate and print the RMSE and R-squared scores on the test set using the predicted and actual values.



```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, predictions))
r2 = r2_score(y_test, predictions)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2) Score: {r2}")
```

## Check for out-of-range predictions

### Subtask:
Determine if any predicted values are outside the valid range for the target variable.


**Reasoning**:
Identify the minimum and maximum quality values in the original dataframe and then check if any of the predictions fall outside this range.



```python
min_quality = df['quality'].min()
max_quality = df['quality'].max()

out_of_range_predictions = predictions[(predictions < min_quality) | (predictions > max_quality)]

if len(out_of_range_predictions) > 0:
    print(f"Out-of-range predictions were found. Number of out-of-range predictions: {len(out_of_range_predictions)}")
else:
    print("No out-of-range predictions were found.")
```

## Improve the model (post-processing)

### Subtask:
Apply post-processing techniques to address out-of-range predictions or further refine the model.


**Reasoning**:
Round the predictions to the nearest integer as wine quality is typically an integer score and display the rounded predictions.



```python
rounded_predictions = np.round(predictions)
display(rounded_predictions)
```

## Re-evaluate the model

### Subtask:
Calculate and report the RMSE and R-squared score on the test set after applying post-processing.


**Reasoning**:
Calculate the RMSE and R-squared score for the rounded predictions and report the results.



```python
rmse_rounded = np.sqrt(mean_squared_error(y_test, rounded_predictions))
r2_rounded = r2_score(y_test, rounded_predictions)

print(f"Root Mean Squared Error (RMSE) after post-processing: {rmse_rounded}")
print(f"R-squared (R2) Score after post-processing: {r2_rounded}")
```
